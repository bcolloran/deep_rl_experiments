```python
basedir = "/home/bc/projects/deep_rl/soft_actor_critic/model_runs/BipedalWalker-v2"
run_dirs = os.listdir(basedir)
complete_run_dirs = list(filter(lambda d: len(os.listdir(basedir+"/"+d))==101, run_dirs))

run_info = []
for rd in complete_run_dirs:  
    with open(f"{basedir}/{rd}/log.pkl", "rb") as pickle_file:
        log_info = pickle.load(pickle_file)
        # print(list(log_info))
        run_info.append((log_info["run params"],log_info["episode log"]))
```

```python
run_summary=[]
for ri in run_info:
    run_summary.append(
        (np.mean(ri[1]["reward"][:100]),
        len(ri[1]["reward"]))
    )
run_summary
```
